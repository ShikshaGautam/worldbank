## Beverages

```python
x=food[['Year', 'Cocoa', 'Coffee_Arabica', 'Coffee_Robusta','Tea']]
y=food['Beverages_avg']
x['Year']=pd.to_datetime(x['Year'])
x['Year']=x.Year.values.astype(np.int64) // 10 ** 9
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()
lnr.fit(x_train,y_train)
py=lnr.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from math import sqrt
print("MSE is:-")
print('Mean Squared Error(MSE):', metrics.mean_squared_error(y_test, py))
```

```python
r=py
ppd = pd.to_datetime(r, unit='s')
x_test['Year']=ppd
```

```python
x_test['food']=y_test
x_test['other']=py
x_test
```

```python
import pandas as pd
import numpy as np
energy=pd.read_csv('energy.csv')
energy
```

```python
energy=energy.rename(columns={'Crude oil, Brent':"Crude_oil_Brent", 'Crude oil, Dubai':"Crude_oil_Dubai", 'Crude oil, WTI':"Crude_oil_WTI", 'Coal, Australian':"Coal_Australian", 'Natural gas, Europe':"Natural_gas_Europe", 'Liquefied natural gas, Japan':"Liquefied_natural_gas_Japan"})
energy
```

```python
energy.columns
```

```python
energy['Crude oil_Brent']=energy['Crude oil_Brent'].str.replace('.','0').astype(float)
energy['Crude oil_WTI']=energy['Crude oil_WTI'].str.replace('.','0').astype(float)
energy['Coal_Australian']=energy['Coal_Australian'].str.replace('.','0').astype(float)
energy['Liquefied natural gas_Japan']=energy['Liquefied natural gas_Japan'].str.replace('.','0').astype(float)
```

## energy

```python
aa=energy[['Year', 'Crude oil_Brent', 'Crude oil_Dubai', 'Crude oil_WTI', 'Coal_Australian', 'Natural gas_Europe', 'Liquefied natural gas_Japan']]
bb=energy['Energy']
aa['Year']=pd.to_datetime(aa['Year'])
aa['Year']=aa.Year.values.astype(np.int64) // 10 ** 9
aa['Crude oil_Brent']=aa['Crude oil_Brent'].astype(np.int64)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(aa,bb,test_size=.20,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr3=LinearRegression()
lr3.fit(x_train,y_train)
pred4=lr3.predict(x_test)
pred4
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from math import sqrt
print("MSE is:-")
print('Mean Squared Error(MSE):', metrics.mean_squared_error(y_test, pred4))
```

```python
r=pred4
ppd2 = pd.to_datetime(r, unit='s')
x_test['Year']=ppd2
```

```python
x_test['energy']=y_test
x_test['other']=pred4
x_test
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.lmplot('energy','other',data=x_test)
plt.show()
```
